```python
%sql
CREATE OR REPLACE TABLE bronze_dev.ruca.ruca_codes AS
SELECT
    PrimaryRuca AS ruca_code,
    PrimaryRucaDescription AS ruca_description
FROM bronze_dev.ruca.ruca_census_tract
GROUP BY
    PrimaryRuca,
    PrimaryRucaDescription;
```

```python
%sql
SELECT * FROM bronze_dev.ruca.ruca_codes ORDER BY 1
```
