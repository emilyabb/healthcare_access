## Shows Connecticut crosswalk

```python
%sql
SELECT DISTINCT
  CountyName20,
  CountyName23,
  CountyFIPS20, 
  CountyFips23
FROM bronze_dev.ruca.ruca_census_tract
WHERE CountyName20 != CountyName23
ORDER BY 1, 2
```

## Create bronze_dev.geo.counties

```python
%sql
CREATE OR REPLACE TABLE bronze_dev.geo.counties AS
SELECT
    CountyName23,
    CountyFIPS23,
    StateName20,
    StateFIPS20
FROM bronze_dev.ruca.ruca_census_tract
GROUP BY ALL
```

```python
%sql
SELECT * FROM bronze_dev.geo.counties ORDER BY StateName20
```
