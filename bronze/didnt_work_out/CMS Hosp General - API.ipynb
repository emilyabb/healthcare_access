Fails with `JSONDecodeError: Expecting value: line 1 column 1 (char 0)`

```python
import requests
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

# API endpoint (latest) 

base_url = "https://data.cms.gov/data-api/v1/dataset/xubh-q36u/data"

# Optional: size (max rows per request) and offset for pagination
# Design pagination as needed to retrieve full dataset
url = f"{base_url}?size=5000&offset=0"

# Fetch JSON with Python (optional)
response = requests.get(url)

print(f"Status: {response.status_code}")
print(f"Content type: {response.headers.get('content-type')}")
print(f"Response length: {len(response.text)}")
print(f"First 200 chars: {response.text[:200]}")
```

```python
import requests
import json

url = "https://data.cms.gov/provider-data/api/1/datastore/sql"

payload = {
    "query": "SELECT * FROM xubh-q36u LIMIT 500 OFFSET 0",
    "show_db_columns": True
}

headers = {
    "Content-Type": "application/json",
    "Accept": "application/json"
}

response = requests.post(url, headers=headers, data=json.dumps(payload))

print(response.status_code)

if response.status_code == 200:
    data = response.json()
    print(f"Rows returned: {len(data)}")
    print(data[:2])  # preview first 2 rows
else:
    print(response.text)
```
