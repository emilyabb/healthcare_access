```python
%sql
CREATE OR REPLACE TABLE silver_dev.geo.counties AS
SELECT
    c.CountyName23 AS county_name,
    c.CountyFIPS23 AS county_fips,
    c.StateName20 AS state_name,
    s.state_abbreviation,
    c.StateFIPS20 AS state_fips,
    CASE WHEN c.StateName20 IN (
        'American Samoa',
        'Commonwealth of the Northern Mariana Islands',
        'Puerto Rico',
        'United States Virgin Islands'
    ) THEN 1 ELSE 0 END AS is_territory,
    
    regexp_replace(
        UPPER(c.CountyName23),
        ' (COUNTY|PARISH|BOROUGH|CENSUS AREA|MUNICIPALITY|MUNICIPIO|CITY AND BOROUGH|PLANNING REGION)$', '') as county_name_no_suffix
FROM
    bronze_dev.geo.counties AS c
LEFT JOIN silver_dev.geo.states AS s ON c.StateName20 = s.state_name
WHERE
    CountyName23 != 'N/A';

SELECT * FROM silver_dev.geo.counties;
```

```python
%sql
SELECT DISTINCT county_name, county_name_no_suffix
FROM silver_dev.geo.counties
```
