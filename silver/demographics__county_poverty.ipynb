```python
%sql
DROP TABLE silver_dev.demographics.poverty_estimates
```

```python
%sql
CREATE OR REPLACE TABLE silver_dev.demographics.county_poverty AS
SELECT
    TO_VARCHAR(FIPS_Code, '999999999') AS county_fips,
    Attribute AS attribute_name,
    
    CASE Attribute
    -- All ages poverty
    WHEN 'POVALL_2023' THEN 'Total number of people in poverty (all ages), 2023'
    WHEN 'CI90LBALL_2023' THEN 'Lower bound of 90% confidence interval for people in poverty (all ages), 2023'
    WHEN 'CI90UBALL_2023' THEN 'Upper bound of 90% confidence interval for people in poverty (all ages), 2023'
    WHEN 'PCTPOVALL_2023' THEN 'Percentage of people in poverty (all ages), 2023'
    WHEN 'CI90LBALLP_2023' THEN 'Lower bound of 90% confidence interval for poverty percentage (all ages), 2023'
    WHEN 'CI90UBALLP_2023' THEN 'Upper bound of 90% confidence interval for poverty percentage (all ages), 2023'

    -- Ages 0–17 poverty
    WHEN 'POV017_2023' THEN 'Number of people age 0–17 in poverty, 2023'
    WHEN 'CI90LB017_2023' THEN 'Lower bound of 90% confidence interval for people age 0–17 in poverty, 2023'
    WHEN 'CI90UB017_2023' THEN 'Upper bound of 90% confidence interval for people age 0–17 in poverty, 2023'
    WHEN 'PCTPOV017_2023' THEN 'Percentage of people age 0–17 in poverty, 2023'
    WHEN 'CI90LB017P_2023' THEN 'Lower bound of 90% confidence interval for poverty percentage age 0–17, 2023'
    WHEN 'CI90UB017P_2023' THEN 'Upper bound of 90% confidence interval for poverty percentage age 0–17, 2023'

    -- Ages 5–17 poverty
    WHEN 'POV517_2023' THEN 'Number of people age 5–17 in poverty, 2023'
    WHEN 'CI90LB517_2023' THEN 'Lower bound of 90% confidence interval for people age 5–17 in poverty, 2023'
    WHEN 'CI90UB517_2023' THEN 'Upper bound of 90% confidence interval for people age 5–17 in poverty, 2023'
    WHEN 'PCTPOV517_2023' THEN 'Percentage of people age 5–17 in poverty, 2023'
    WHEN 'CI90LB517P_2023' THEN 'Lower bound of 90% confidence interval for poverty percentage age 5–17, 2023'
    WHEN 'CI90UB517P_2023' THEN 'Upper bound of 90% confidence interval for poverty percentage age 5–17, 2023'

    -- Median household income
    WHEN 'MEDHHINC_2023' THEN 'Median household income, 2023'
    WHEN 'CI90LBINC_2023' THEN 'Lower bound of 90% confidence interval for median household income, 2023'
    WHEN 'CI90UBINC_2023' THEN 'Upper bound of 90% confidence interval for median household income, 2023'

    -- Ages 0–4 poverty
    WHEN 'POV04_2023' THEN 'Number of people age 0–4 in poverty, 2023'
    WHEN 'CI90LB04_2023' THEN 'Lower bound of 90% confidence interval for people age 0–4 in poverty, 2023'
    WHEN 'CI90UB04_2023' THEN 'Upper bound of 90% confidence interval for people age 0–4 in poverty, 2023'
    WHEN 'PCTPOV04_2023' THEN 'Percentage of people age 0–4 in poverty, 2023'
    WHEN 'CI90LB04P_2023' THEN 'Lower bound of 90% confidence interval for poverty percentage age 0–4, 2023'
    WHEN 'CI90UB04P_2023' THEN 'Upper bound of 90% confidence interval for poverty percentage age 0–4, 2023'

    -- Rural/Urban classifications
    WHEN 'Rural_Urban_Continuum_Code_2013' THEN 'USDA Rural-Urban Continuum Code, 2013'
    WHEN 'Urban_Influence_Code_2013' THEN 'USDA Urban Influence Code, 2013'
    WHEN 'Rural_Urban_Continuum_Code_2023' THEN 'USDA Rural-Urban Continuum Code, 2023'
    WHEN 'Urban_Influence_Code_2024' THEN 'USDA Urban Influence Code, 2024'

    ELSE 'Unknown attribute'
    END AS attribute_description,
    Value AS attribute_value
FROM
    bronze_dev.economic_research_service.county_poverty;
```
