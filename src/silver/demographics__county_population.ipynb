```python

%sql
CREATE OR REPLACE TABLE silver_dev.census_bureau.county_population AS
SELECT
  counties.county_fips,
  counties.state_fips,
  pop.pop_base_2020 AS population_base_2020,
  pop.pop_est_2020 AS population_estimate_2020,
  pop.pop_est_2021 AS population_estimate_2021,
  pop.pop_est_2022 AS population_estimate_2022,
  pop.pop_est_2023 AS population_estimate_2023,
  pop.pop_est_2024 AS population_estimate_2024,
  pop.pop_est_2025 AS population_estimate_2025
FROM
  bronze_dev.census_bureau.county_population AS pop
LEFT JOIN silver_dev.geo.counties AS counties
  ON pop.county_name = counties.county_name
  AND pop.state_name = counties.state_name;
```
