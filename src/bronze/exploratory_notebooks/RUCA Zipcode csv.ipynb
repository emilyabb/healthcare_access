```python

file_path_csv = "/Volumes/bronze_dev/ruca/ruca_raw/RUCA-codes-2020-zipcode.csv"

df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(file_path_csv)

display(df)
```

```python
from pyspark.sql.functions import col, lpad

df = df.withColumn(
    "ZIPCode",
    lpad(col("ZIPCode").cast("string"), 5, "0")
)

display(df)
```

```python
df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze_dev.ruca.ruca_zipcode")
```
