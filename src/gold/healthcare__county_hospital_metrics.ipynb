```python
%sql
CREATE OR REPLACE TABLE gold_dev.healthcare.county_hospital_summary
SELECT
    county_fips,
    AVG(hospitals.hospital_overall_rating) AS avg_hospital_rating,
    COUNT(*) AS ct_hospitals
FROM
  silver_dev.healthcare.hospitals AS hospitals
WHERE
  county_fips IS NOT NULL
GROUP BY
  hospitals.county_fips
```
