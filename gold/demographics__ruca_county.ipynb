```python
%sql
CREATE OR REPLACE TABLE gold_dev.demographics.ruca_county
SELECT
	county_fips_23 AS county_fips,
	AVG(primary_ruca * 1.0) AS avg_primary_ruca,
	SUM(land_area) AS land_area,
	TRY_DIVIDE(
    SUM(population) * 1.0,
    SUM(land_area)
   ) AS population_density	
FROM
	silver_dev.demographics.ruca_census_tract
GROUP BY
	county_fips_23;
```
